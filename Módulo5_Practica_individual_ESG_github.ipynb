## Imputación de las demás variables cualitativas del dataframe con su moda correspondiente

```python
#Dataframe
copy_ind_risk.info()
```

```python
#Eliminación de la columna 'Symbol'
copy_ind_risk.drop(columns=['Symbol'])
copy_ind_risk.info()
```

```python
def imputar_moda(copy_ind_risk, variables_cualitativas):

  for variable in variables_cualitativas:
    moda_variable = copy_ind_risk[variable].mode()[0]
    copy_ind_risk[variable].fillna(moda_variable, inplace=True)

  return copy_ind_risk
```

```python
variables_cualitativas = list(copy_ind_risk.select_dtypes('object').columns)
```

```python
copy_ind_risk = imputar_moda(copy_ind_risk = copy_ind_risk,
                      variables_cualitativas = variables_cualitativas
                      )
```

```python
copy_ind_risk.info()
```

## Imputación de las variables cuantitativas del dataframe con la moda de cada variable

```python
def impute_with_mode_col(risk_8, variables_cuantitativas):
    for column in variables_cuantitativas:
            mode_value = risk_8[column].mode()[0]
            risk_8[column].fillna(mode_value, inplace=True)
    return risk_8
```

```python
variables_cuantitativas = list(copy_ind_risk.select_dtypes('float64').columns)
```

```python
risk_8 = impute_with_mode_col(risk_8 = copy_ind_risk,
                          variables_cuantitativas = variables_cuantitativas
                          )
```

```python
risk_8.info()
```

mae_test y r2_test:

mae_test: Mean Absolute Error (Error Absoluto Medio) en datos de prueba. Es una métrica que mide el promedio de las diferencias absolutas entre las predicciones del modelo y los valores reales en el conjunto de datos de prueba. En este caso, el valor es aproximadamente 0.6005. Un valor más bajo indica que el modelo tiene un error absoluto medio más pequeño, lo que significa que las predicciones son más cercanas a los valores reales.

r2_test: Coeficiente de determinación (R-cuadrado) en datos de prueba. Es una métrica que indica cuánta variabilidad en los datos de prueba es explicada por el modelo. El valor es aproximadamente 0.9873. Un valor cercano a 1 sugiere que el modelo se ajusta muy bien a los datos de prueba y es capaz de explicar la mayor parte de la variabilidad en los valores observados.

mae_train y r2_train:

mae_train: Mean Absolute Error (Error Absoluto Medio) en datos de entrenamiento. Es similar a mae_test, pero se calcula en el conjunto de datos de entrenamiento utilizado para entrenar el modelo. En este caso, el valor es aproximadamente 0.5401.

r2_train: Coeficiente de determinación (R-cuadrado) en datos de entrenamiento. Al igual que r2_test, indica cuánta variabilidad en los datos de entrenamiento es explicada por el modelo. En este caso, el valor es aproximadamente 0.9902.
Estos resultados indican que el modelo tiene un rendimiento muy bueno en ambos conjuntos de datos (entrenamiento y prueba). El valor bajo de mae indica que las predicciones del modelo están muy cerca de los valores reales, y el valor alto de r2 sugiere que el modelo es altamente capaz de explicar la variabilidad en los datos. Sin embargo, es importante tener en cuenta que un alto rendimiento en el conjunto de entrenamiento (como se muestra en r2_train) a veces puede ser un indicio de sobreajuste (overfitting), por lo que es importante evaluar el modelo en datos no vistos (conjunto de prueba) para verificar su generalización.
