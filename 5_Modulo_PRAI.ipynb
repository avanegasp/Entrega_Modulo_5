## Imputación de las demás variables cualitativas del dataframe con su moda correspondiente

```python
#Dataframe
copy_ind_risk.info()
```

```python
#Eliminación de la columna 'Symbol'
copy_ind_risk.drop(columns=['Symbol'])
copy_ind_risk.info()
```

```python
def imputar_moda(copy_ind_risk, variables_cualitativas):

  for variable in variables_cualitativas:
    moda_variable = copy_ind_risk[variable].mode()[0]
    copy_ind_risk[variable].fillna(moda_variable, inplace=True)

  return copy_ind_risk
```

```python
variables_cualitativas = list(copy_ind_risk.select_dtypes('object').columns)
```

```python
copy_ind_risk = imputar_moda(copy_ind_risk = copy_ind_risk,
                      variables_cualitativas = variables_cualitativas
                      )
```

```python
copy_ind_risk.info()
```

## Imputación de las variables cuantitativas del dataframe con la moda de cada variable

```python
def impute_with_mode_col(risk_8, variables_cuantitativas):
    for column in variables_cuantitativas:
            mode_value = risk_8[column].mode()[0]
            risk_8[column].fillna(mode_value, inplace=True)
    return risk_8
```

```python
variables_cuantitativas = list(copy_ind_risk.select_dtypes('float64').columns)
```

```python
risk_8 = impute_with_mode_col(risk_8 = copy_ind_risk,
                          variables_cuantitativas = variables_cuantitativas
                          )
```

```python
risk_8.info()
```

## Mean Absolute Error (Error Absoluto Medio) y Coeficiente de determinación (R-cuadrado)

```python
from sklearn.metrics import mean_absolute_error, r2_score

X = risk_8[['Environment Risk Score', 'Governance Risk Score', 'Social Risk Score']]

y = risk_8['Total ESG Risk score']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 99)


modelo_1 = LinearRegression()
modelo_1.fit(X_train, y_train)

y_predict_test = modelo_1.predict(X_test)
y_predict_train = modelo_1.predict(X_train)

mae_test = mean_absolute_error(y_test, y_predict_test)
r2_test = r2_score(y_test, y_predict_test)

mae_train = mean_absolute_error(y_train, y_predict_train)
r2_train = r2_score(y_train, y_predict_train)

print('Prueba - Error Absoluto Medio:',round(mae_test,2), 'y Coeficiente de determinación:', round(r2_test,2))

print('Entrenamiento - Error Absoluto Medio:', round(mae_train,2), 'y Coeficiente de determinación:', round(r2_train,2))
```

**Coeficiente de determinación (R-cuadrado) en datos de prueba.** El valor es aproximadamente 0.9873. Un valor cercano a 1 sugiere que el modelo se ajusta muy bien a los datos de prueba y es capaz de explicar la mayor parte de la variabilidad en los valores observados.
